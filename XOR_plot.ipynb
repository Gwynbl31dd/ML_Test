XOR plot

```
import numpy as np;
import matplotlib.pyplot as plt;
```

```
x = np.random.randint(-100,100,10000)
y = np.random.randint(-100,100,10000)
z = np.bitwise_xor(x,y)  
# print(z)
#y = np.zeros(100);
#y[:50] = 1;

plt.scatter(x,y,c=z);
```
